```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import quad
```

We consider the effects of a diagonal metrix d$s^2$ = $g_{00}$d$t^2$ + $g_{11}$d$x_1^2$ + $g_{22}$d$x_2^2$ + $g_{33}$d$x_3^2$ on a cold two-stream instability. The two streams move in the $x_1$-direction with velocity $v_0$ and $-v_0$. At each point, we can choose local coordinates $\tilde{t} = \sqrt{-g_{00}} t$ and $\tilde{x}_1 = \sqrt{g_{11}} x_1$ such that, in the frame of these local coordinates, spacetime is locally flat. From the perspective of an observer at infinity, the flow of time at the point $\tilde{x}_\mu$ is altered by a factor $\sqrt{-g_{00}(\tilde{x}_\mu)}$. We plot this factor for a Schwarzschild black hole in function of $r$. This effect is gravitational time dilation, and it alters the speed at which instabilities in the two-stream instability grow.

```python
r_range = np.linspace(1, 10, 5000)
g00_range = -1 + 1/r_range

plt.figure()
plt.plot(r_range, np.sqrt(-g00_range))
plt.xlim([0, 10])
plt.ylim([0, 1])
plt.xlabel("r  [2GM/c$^2$]")
plt.ylabel("$\\sqrt{-g_{00}}$")
plt.title("Gravitation time dilation in a Schwarzschild metric")
plt.show()
```

The length scale at which the instabilities occur depends on the citical length scale $\tilde{k}_\text{crit} = k_\text{crit} / \sqrt{g_{11}}$, which is determined by the formula $\tilde{k}^2_\text{crit} \tilde{v}_0^2 \tilde{\gamma}_0^3 = \tilde{\omega}_{p,0}^2$, where $\tilde{\omega}_{p,0}$ is the plasma frequency in the locally flat coordinates. This formula which is equivalent to $k^2_\text{crit} v_0^2 \tilde{\gamma}_0^3 = \omega_{p,0}^2$. Here we see that the metric alters the critical length scale $k_\text{crit}$ only through the Lorentz factor. Compared to special relativity,  $k_\text{crit}$ picks up an extra factor $\tilde{\gamma}_0^3 / \gamma_0^3 = \sqrt{1 - v_0^2} / \sqrt{1 - \frac{g_{11}}{-g_{00}} v_0^2}$. We plot this factor for several values of $v_0$.

```python
plt.figure()
for v0 in [0.1, 0.3, 0.5, 0.7, 0.9]:
    gamma = np.sqrt(1 - v0**2)
    gamma_tilde_range = np.sqrt(1 - v0**2 / (g00_range)**2)
    factor = gamma_tilde_range**3 / gamma**3

    plt.plot(r_range, factor, label="$v_0 = $"+str(v0)+"c")
plt.xlim([1, 10])
plt.ylim([0, 1])
plt.xlabel("r  [2GM/c$^2$]")
plt.ylabel("$(\\tilde{\\gamma}_0 / \\gamma_0)^3$")
plt.title("Modification factor of $k_\\text{crit}$ in a Schwarzschild metric")
plt.legend()
plt.show()
```

We find that, closer to the black hole, the instabilities form at larger scales. Notice how, given a speed $v_0$, there are some coordinates $r$ close to the black hole at which $k_\text{crit}$ is not well-defined. In order for the apparant speed of the two streams to be $v_0$ at the location $x_\mu$, the streams must have a speed $\tilde{v}_0 = \sqrt{\frac{g_{11}(x_\mu)}{-g_{00}(x_\mu)}} v_0$ in the locally flat frame at $x_\mu$. This speed $\tilde{v}_0$ must not exceed the speed of light. Therefore, the speed $|v_0|$ cannot exceed $\sqrt{\frac{-g_{00}}{g_{11}}}$. The following figure shows which parts of phase space are accessible due to this constraint.

```python
plt.figure()
plt.fill_between(r_range, g00_range, -g00_range, alpha=0.5)
plt.xlim([1, 10])
plt.ylim([-1, 1])
plt.xlabel("r  [2GM/c$^2$]")
plt.ylabel("v  [c]")
plt.title("Accessible region of phase space")
plt.show()
```

We can also plot the factor $\tilde{\gamma}_0^3 / \gamma_0^3$ for different values of $\tilde{v}_0$. In the limit of $r \to 1$, the modification factor of $k_\text{crit}$ becomes $(1 - \tilde{v}_0^2)^{3/2}$.

```python
plt.figure()
for v0 in [0.1, 0.3, 0.5, 0.7, 0.9]:
    gamma = np.sqrt(1 - v0**2 * (g00_range)**2)
    gamma_tilde_range = np.sqrt(1 - v0**2)
    factor = gamma_tilde_range**3 / gamma**3

    plt.plot(r_range, factor, label="$\\tilde{v}_0 = $"+str(v0)+"c")
plt.xlim([1, 10])
plt.ylim([0, 1])
plt.xlabel("r  [2GM/c$^2$]")
plt.ylabel("$(\\tilde{\\gamma}_0 / \\gamma_0)^3$")
plt.title("Modification factor of $k_\\text{crit}$ in a Schwarzschild metric")
plt.legend()
plt.show()
```

As the two-stream stability evolves, the gravity of the black hole will cause the speed and density of both streams to change. In general, it will no longer be true that the two streams have the same speed but oppositely oriented. At each position $x_\mu$, it will still be possible to change to a frame of reference where the streams have speeds $v_0$ and $-v_0$, but in general the densities of the streams will still be different. As a result, the two-stream instability will no longer adhere to the formula $\tilde{k}^2_\text{crit} \tilde{v}_0^2 \tilde{\gamma}_0^3 = \tilde{\omega}_{p,0}^2$.
